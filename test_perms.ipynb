```python
from models import db, User, Student
from app import app

email = "yourEmail@colby.edu"
```

## Check current role

```python
with app.app_context():
    peter = User.query.filter_by(email=email).first()

    print(peter.role)
```

## Make admin

```python
with app.app_context():
    peter = User.query.filter_by(email=email).first()

    peter.role = 'admin'
    db.session.commit()
```

## Make admin manager

```python
with app.app_context():
    peter = User.query.filter_by(email=email).first()

    peter.role = 'admin_manager'
    db.session.commit()
```

## Make None

```python
with app.app_context():
    peter = User.query.filter_by(email=email).first()

    peter.role = None
    db.session.commit()
```

## Make student

```python
with app.app_context():
    peter = User.query.filter_by(email=email).first()

    peter.role = 'student'
    db.session.commit()
```

## Add self to student table

```python
with app.app_context():
    new_student = Student(
        student_id=999999,
        email=email,
        first_name='Test',
        last_name='User',
    )

    db.session.add(new_student)
    db.session.commit()
```

## Remove self from student table

```python
with app.app_context():
    student = Student.query.filter_by(email=email).first()
    db.session.delete(student)
    db.session.commit()
```
